## 5.1. Актуальные данные по гостиницам

#### Финальный список гостиниц, в котором по каждой гостинице - только одна запись с самыми актуальными и полными данными

```python
agg_data1 = data1.groupby('title')['create_time'].max().reset_index()

res_data1 = pd.merge(data1, agg_data1, on=['title', 'create_time'], how='inner')
res_data1
```

## 5.3. Квадрат координат размером 1км на 1км, где больше всего гостиниц

### 5.3.1. Определим минимальное и максимальное значение координат:

```python
min_lat = res_data1['lat'].min()
max_lat = res_data1['lat'].max()

min_lon = res_data1['lon'].min()
max_lon = res_data1['lon'].max()
```

### 5.3.2. Разделим диапазон координат на квадраты размером 1 км:

```python
step = 1 / 111
sq_lat = pd.cut(res_data1['lat'], bins=int((max_lat-min_lat)/step), include_lowest=True)
sq_lon = pd.cut(res_data1['lon'], bins=int((max_lon-min_lon)/step), include_lowest=True)
```

### 5.3.3. Сгруппируем датасет по комбинации квадратов координат и посчитаем количество гостиниц в каждой группе:

```python
grouped = res_data1.groupby([sq_lat, sq_lon]).size().reset_index(name='количество_гостиниц')
```

### 5.3.4. Квадрат координат с максимальным количеством гостиниц

```python
max_count_square = grouped[grouped['количество_гостиниц'] == grouped['количество_гостиниц'].max()]
```

```python
print('Координаты квадрата с наибольшим количеством гостиниц:')
print('lat:', max_count_square['lat'].values[0])
print('lot:', max_count_square['lon'].values[0])
```

## 5.4. Карта

```python
res_data_sq1 = res_data1.query('(lat >= 54.709 and lat <= 54.718) and (lon >= 20.499 and lon <= 20.508)')

shir = 54.714
dol = 20.504

karta = folium.Map(location=[shir, dol], zoom_start=5)

def create(row):
    Marker([row['lat'], row['lon']], popup=f"{row['title']}").add_to(karta)
    
res_data_sq1.apply(create, axis=1)

karta
```
